```
import sys
sys.path.insert(0, r'C:\\mcop\\src')
import numpy as np
import pandas as pd
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
from gbm_simulator import simulate_gbm
from option_pricer import price_european, price_asian, price_barrier
from risk_metrics import calculate_var_cvar, stress_test

S0=100; K=105; B=85; r=0.05; sigma=0.20; T=1.0; steps=252; n_sim=50000
print('All imports successful | Parameters set')
```

```
paths = simulate_gbm(S0, r, sigma, T, steps, n_sim=200)
plt.figure(figsize=(12,5))
plt.plot(paths[:,:50], alpha=0.3, linewidth=0.7)
plt.axhline(y=K, color='red', linestyle='--', label=f'Strike K={K}')
plt.axhline(y=B, color='orange', linestyle='--', label=f'Barrier B={B}')
plt.title('Monte Carlo GBM Simulated Price Paths')
plt.xlabel('Trading Days'); plt.ylabel('Asset Price')
plt.legend(); plt.tight_layout()
plt.savefig(r'C:\\mcop\\outputs\\gbm_paths.png', dpi=150)
plt.close()
print('GBM chart saved')
```

```
eu_call, eu_se = price_european(S0,K,r,sigma,T,steps,n_sim,'call')
eu_put,  _    = price_european(S0,K,r,sigma,T,steps,n_sim,'put')
asian,   _    = price_asian(S0,K,r,sigma,T,steps,n_sim,'call')
bar_do,  _    = price_barrier(S0,K,B,r,sigma,T,steps,n_sim,'down-and-out')
bar_di,  _    = price_barrier(S0,K,B,r,sigma,T,steps,n_sim,'down-and-in')

results = pd.DataFrame({
    'Option Type': ['European Call','European Put','Asian Call','Barrier Down-Out','Barrier Down-In'],
    'MC Price'   : [round(eu_call,4),round(eu_put,4),round(asian,4),round(bar_do,4),round(bar_di,4)]
})
print(results.to_string(index=False))
```

```
var95,cvar95,returns = calculate_var_cvar(S0,r,sigma,T,steps,n_sim,0.95)
var99,cvar99,_       = calculate_var_cvar(S0,r,sigma,T,steps,n_sim,0.99)

plt.figure(figsize=(11,5))
plt.hist(returns, bins=100, alpha=0.7, color='steelblue', edgecolor='white')
plt.axvline(var95,  color='orange', linestyle='--', linewidth=2, label=f'VaR  95%: {var95:.2%}')
plt.axvline(cvar95, color='red',    linestyle='--', linewidth=2, label=f'CVaR 95%: {cvar95:.2%}')
plt.axvline(var99,  color='purple', linestyle='--', linewidth=2, label=f'VaR  99%: {var99:.2%}')
plt.title('Return Distribution - VaR & CVaR')
plt.xlabel('Return'); plt.ylabel('Frequency')
plt.legend(); plt.tight_layout()
plt.savefig(r'C:\\mcop\\outputs\\var_cvar.png', dpi=150)
plt.close()
print(f'95% VaR:{var95:.2%} | 95% CVaR:{cvar95:.2%}')
print(f'99% VaR:{var99:.2%} | 99% CVaR:{cvar99:.2%}')
```

```
base, stress = stress_test(S0,K,r,sigma,T,steps,n_sim)
stress_df = pd.DataFrame(stress).T
print(f'Base Option Price: {base:.4f}')
print(stress_df.to_string())

stress_df['P&L Change (%)'].plot(
    kind='bar',
    color=['#e74c3c','#c0392b','#922b21','#641e16'],
    figsize=(8,4),
    title='Stress Test: P&L Change Under Market Shocks'
)
plt.ylabel('P&L Change (%)'); plt.xticks(rotation=0); plt.tight_layout()
plt.savefig(r'C:\\mcop\\outputs\\stress_test.png', dpi=150)
plt.close()
print('All outputs saved to C:\\mcop\\outputs\\')
```
